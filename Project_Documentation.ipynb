## 📊 Language Composition Visualization

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Language composition data
languages = ['Python', 'HTML']
percentages = [50.1, 49.9]
colors = ['#3776ab', '#e34c26']

# Create pie chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Pie Chart
wedges, texts, autotexts = ax1.pie(percentages, labels=languages, autopct='%1.1f%%',
                                    colors=colors, startangle=90, textprops={'fontsize': 12})
ax1.set_title('Language Composition', fontsize=14, fontweight='bold', pad=20)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# Bar Chart
bars = ax2.bar(languages, percentages, color=colors, edgecolor='black', linewidth=1.5)
ax2.set_ylabel('Percentage (%)', fontsize=12)
ax2.set_title('Language Distribution', fontsize=14, fontweight='bold', pad=20)
ax2.set_ylim(0, 100)
for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Display as table
lang_df = pd.DataFrame({
    'Language': languages,
    'Percentage': percentages,
    'Primary Use': ['Backend, ML Models, APIs', 'Frontend, Dashboard, UI']
})
print("\n" + "="*70)
print("Language Distribution Summary")
print("="*70)
print(lang_df.to_string(index=False))
print("="*70)
````

## 🔧 Core Components Detailed Documentation

### 1. Demand Forecasting Module

- Historical sales data

- Market indicators

- Demand forecast with confidence intervals
- Trend analysis

**Example Usage**:


---

### 2. Warehouse Optimizer

**Location**: `/models/warehouse_optimizer.py`  
**Purpose**: Select optimal warehouse for order fulfillment

**Input Parameters**:
- Order details (product ID, quantity)
- Warehouse inventory levels
- Location coordinates
- Delivery date requirement

**Output**:
- Best warehouse recommendation
- Cost analysis
- Delivery time estimation
- Alternative warehouse suggestions

**Optimization Criteria**:
- Minimum transportation cost
- Shortest delivery distance
- Inventory availability
- Warehouse capacity
- Delivery date feasibility

**Example Usage**:
```python

optimizer = WarehouseOptimizer()
best = optimizer.select_warehouse(order_details)
```

---

### 3. Route Optimizer

**Location**: `/models/route_optimizer.py`  
**Purpose**: Calculate most efficient delivery routes

**Input Parameters**:
- Multiple delivery points
- Traffic data and conditions
- Vehicle constraints
- Time windows for delivery

**Output**:
- Optimized route with waypoints
- Estimated time and cost
- Vehicle assignment
- Alternative routes

**Algorithm**: Vehicle Routing Problem (VRP) solver using metaheuristics

**Example Usage**:
```python
from models.route_optimizer import RouteOptimizer
optimizer = RouteOptimizer()
route = optimizer.optimize(delivery_points)
```

**Output**:

## 📈 Usage Examples & Code Snippets

### Example 1: Complete Demand Forecasting Workflow

````python
# Example: Demand Forecasting Workflow
example_code_1 = """
from models.demand_forecast import DemandForecaster
import pandas as pd
import matplotlib.pyplot as plt

# Initialize forecaster
forecaster = DemandForecaster(model='prophet')

# Load historical data
data = pd.read_csv('data/sales_history.csv')

# Train the model
forecaster.fit(data)

# Make forecast
forecast = forecaster.predict(
    product_id='PROD001',
    periods=30,  # 30 days ahead
    confidence_level=0.95
)

# Display results
print(f"Forecasted Demand (30 days): {forecast['mean']:.2f} units")
print(f"95% Confidence Interval: [{forecast['lower']:.2f}, {forecast['upper']:.2f}]")

# Visualize
forecast.plot()
plt.show()
"""

print("📊 DEMAND FORECASTING EXAMPLE:")
print("="*60)
print(example_code_1)
````

### Example 2: Warehouse Selection Algorithm

````python
# Example: Warehouse Selection
example_code_2 = """
from models.warehouse_optimizer import WarehouseOptimizer
from datetime import datetime, timedelta

# Initialize optimizer
optimizer = WarehouseOptimizer()

# Define order
order = {
    'order_id': 'ORD123456',
    'product_id': 'PROD001',
    'quantity': 50,
    'delivery_location': {
        'lat': 40.7128,
        'lon': -74.0060,
        'address': 'New York, NY'
    },
    'delivery_date': (datetime.now() + timedelta(days=5)).strftime('%Y-%m-%d')
}

# Get best warehouse
best_warehouse = optimizer.select_warehouse(order)

# Display results
print(f"Best Warehouse: {best_warehouse['name']}")
print(f"Location: {best_warehouse['location']}")
print(f"Estimated Cost: ${best_warehouse['cost']:.2f}")
print(f"Estimated Delivery Time: {best_warehouse['days']} days")
print(f"Inventory Available: {best_warehouse['stock']} units")
"""

print("🏭 WAREHOUSE SELECTION EXAMPLE:")
print("="*60)
print(example_code_2)
````

### Example 3: Route Optimization

````python
# Example: Route Optimization
example_code_3 = """
from models.route_optimizer import RouteOptimizer
import json

# Initialize optimizer
optimizer = RouteOptimizer()

# Define delivery points
delivery_points = [
    {'id': 1, 'name': 'Customer A', 'lat': 40.7128, 'lon': -74.0060, 'volume': 10},
    {'id': 2, 'name': 'Customer B', 'lat': 40.7580, 'lon': -73.9855, 'volume': 15},
    {'id': 3, 'name': 'Customer C', 'lat': 40.7489, 'lon': -73.9680, 'volume': 20},
    {'id': 4, 'name': 'Customer D', 'lat': 40.7505, 'lon': -73.9972, 'volume': 12}
]

# Optimize route
optimized_route = optimizer.optimize(delivery_points)

# Display results
print("\nOptimized Route:")
for i, point in enumerate(optimized_route['route'], 1):
    print(f"{i}. {point['name']} - {point['lat']}, {point['lon']}")

print(f"\nTotal Distance: {optimized_route['distance']:.2f} km")
print(f"Total Time: {optimized_route['time']:.2f} hours")
print(f"Estimated Cost: ${optimized_route['cost']:.2f}")
"""

print("🚚 ROUTE OPTIMIZATION EXAMPLE:")
print("="*60)
print(example_code_3)
````
